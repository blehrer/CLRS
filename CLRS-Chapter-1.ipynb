## Exercises
### 1.2-2
Suppose we are comparing implementations of insertion sort and merge sort on the same machine. For inputs of size n, insertion sort runs in `8n^2` steps, and merge sort runs in `64nlog(n)` steps. For which values of `n` does insertion sort beat merge sort?
> n = [2,3,]
> it does not take long for `O(nlog(n))` to be faster than `O(n^2)`

```python


import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 10, 10)  # Sample data.

fig, ax = plt.subplots(layout='constrained')
ax.plot(x, 64 * np.log2(x), label='merge-sort, 64*log₂(x)')  # Plot some data on the Axes.
ax.plot(x, 8 * x ** 2, label='insert-sort, 8 * x²')  # Plot more data on the Axes...
ax.set_xlabel('# of inputs')  # Add an x-label to the Axes.
ax.set_ylabel('# of steps')  # Add a y-label to the Axes.
ax.set_title("Irrelevance of coefficients at scale")  # Add a title to the Axes.
ax.legend(title="Time-complexity Signatures")  # Add a legend.
```

### 1.2-3
What is the smallest value of `n` such than an algorithm whose running time is `100n^2` runs faster than an algorithm whose running time  is `2^n` on the same machine?
> n = 15
> It does not take long for `O(n^2)` to be faster than `O(2^n)`

```python
x = np.linspace(1, 30, 30)  # Sample data.
fig, ax = plt.subplots(layout='constrained')
nsquared = 100 * x * x
two_to_the_n = np.pow(2, x)
ax.plot(x, nsquared, label='100n²')  # Plot some data on the Axes.
ax.plot(x, two_to_the_n, label='2ⁿ', )  # Plot more data on the Axes...
ax.set_xlabel('# of inputs')  # Add an x-label to the Axes.
ax.set_yscale('log', base=2)
ax.axvline(14, color='red', label='tradeoff, favor 2ⁿ')
ax.axvline(15, color='lightgreen', label='tradeoff, favor 100n²')
ax.set_ylabel('# of steps')  # Add a y-label to the Axes.
ax.set_title("Trade-offs")  # Add a title to the Axes.
ax.legend(title="Time-complexity Signatures")  # Add a legend.
```
