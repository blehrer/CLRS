## block comment with tex
Put `> $$`, then next lines; end with `$$` ... not the same as usual, where `>` precedes every line.
```markdown
> $$
\begin{flalign}
\mathrm{T}(n) = \begin{cases}
    2 & \text{if } n = 2 \\
    n\lg{n} & \text{if } 2^k \text{, for } k>1
\end{cases}&&
\end{flalign}
$$
```

> $$
\begin{flalign}
\mathrm{T}(n) = \begin{cases}
    2 & \text{if } n = 2 \\
    n\lg{n} & \text{if } 2^k \text{, for } k>1
\end{cases}&&
\end{flalign}
$$
